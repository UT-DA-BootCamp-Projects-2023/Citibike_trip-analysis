## Dependencies and Setup

```python
# import required libiraries/packages

import pandas as pd
from pathlib import Path

```

```python
# define citbike data file name list
citibikefilelist = ['JC-202206-citibike-tripdata.csv',
                'JC-202207-citbike-tripdata.csv', 
                'JC-202208-citibike-tripdata.csv',
                'JC-202209-citibike-tripdata.csv',
                'JC-202210-citibike-tripdata.csv',
                'JC-202211-citibike-tripdata.csv',
                'JC-202212-citibike-tripdata.csv',
                'JC-202301-citibike-tripdata.csv',
                'JC-202302-citibike-tripdata.csv',
                'JC-202303-citibike-tripdata.csv',
                'JC-202304-citibike-tripdata.csv',
                'JC-202305-citibike-tripdata.csv']
# Create Empty DataFrame with Column Header
citibike_jun22_to_may23_df = pd.DataFrame(columns=['ride_id', 'rideable_type', 'started_at', 'ended_at','start_station_name',
                                                   'start_station_id','end_station_name','end_station_id','start_lat','start_lng',
                                                   'end_lat','end_lng','member_casual'])

# Import citibike-tripdata 
for eachfileName in citibikefilelist:
    filepath = Path(eachfileName)
    tempDF = pd.read_csv(filepath,encoding="utf-8")

    citibike_jun22_to_may23_df = pd.concat([citibike_jun22_to_may23_df,tempDF],ignore_index=True)


# Display dataframe
citibike_jun22_to_may23_df
```

```python
# Remove unwanted column name 'A'
citibike_jun22_to_may23_df = citibike_jun22_to_may23_df.drop(['start_station_id','end_station_id'], axis=1)
```

```python
#  Check for null values in the dataframe under an entire DataFrame:
citibike_jun22_to_may23_df.loc[citibike_jun22_to_may23_df.isna().any(axis=1)]
```

```python
# remove null value data
citibike_jun22_to_may23_df = citibike_jun22_to_may23_df.dropna().reset_index(drop=True)
citibike_jun22_to_may23_df
```

```python
# Check if ride ID column as unique values
citibike_jun22_to_may23_df['ride_id'].is_unique
```

```python
# Export the data to CSV
citibike_jun22_to_may23_df.to_csv("citibike_jun22_to_may23_df.csv",index=False)
```
